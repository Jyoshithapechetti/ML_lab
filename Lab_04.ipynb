```python
#A4
import pandas as pd  # Import the pandas library for data manipulation and analysis.
import numpy as np   # Import the numpy library for numerical operations.

# Specify the file path to the Excel file containing the data.
file_path = 'Lab Session Data.xlsx'

# Read the Excel file into a DataFrame. The specific sheet being read is 'IRCTC Stock Price'.
df = pd.read_excel(file_path, sheet_name='IRCTC Stock Price')

# Display the first few rows of the DataFrame to get a quick overview of the data.
df.head()
```

```python
mean_price_population=df['Price'].mean()
print(f"Mean = {mean_price_population}")
print('Variance = ', df['Price'].var())

#extracting all wednesdays
wed_df = df[df['Day'] == 'Wed']

# Calculate the mean of the 'Price' column for the filtered DataFrame
mean_price_wed = wed_df['Price'].mean()

# Display the result
print(f"Mean Price on Wednesdays: {mean_price_wed}")
```

```python
#Comparing population and wednesday mean
# Compare the mean stock price on Wednesdays with the overall population mean price.
if mean_price_wed > mean_price_population:
    # If the mean price on Wednesdays is higher than the overall mean price, print this message.
    print("The mean price on Wednesdays is higher than the overall population mean price.")
    
elif mean_price_wed < mean_price_population:
    # If the mean price on Wednesdays is lower than the overall mean price, print this message.
    print("The mean price on Wednesdays is lower than the overall population mean price.")
    
else:
    # If the mean price on Wednesdays is equal to the overall mean price, print this message.
    print("The mean price on Wednesdays is equal to the overall population mean price.")
```

```python
#A8 continution
import numpy as np

# Calculate JC and SMC
a11 = np.sum((vector1_binary == 1) & (vector2_binary == 1))
a00 = np.sum((vector1_binary == 0) & (vector2_binary == 0))
a01 = np.sum((vector1_binary == 0) & (vector2_binary == 1))
a10 = np.sum((vector1_binary == 1) & (vector2_binary == 0))

# Debugging: Print values of a11, a00, a01, and a10
print(f"a11: {a11}, a00: {a00}, a01: {a01}, a10: {a10}")

# Compute Jaccard Coefficient (JC)
denominator_jc = (a01 + a10 + a11)
if denominator_jc != 0:
    jc = a11 / denominator_jc
else:
    jc = np.nan  # or some default value if you prefer

# Compute Simple Matching Coefficient (SMC)
denominator_smc = (a00 + a01 + a10 + a11)
if denominator_smc != 0:
    smc = (a11 + a00) / denominator_smc
else:
    smc = np.nan  # or some default value if you prefer

print(f"Jaccard Coefficient (JC): {jc}")
print(f"Simple Matching Coefficient (SMC): {smc}")
```

```python
#A10
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

# Select the first 20 rows from the DataFrame
subset_data = data.head(20)

# Convert categorical variables to numerical and handle any missing values
binary_data = subset_data.copy()
for column in binary_data.columns:
    if binary_data[column].dtype == 'object':
        encoder = LabelEncoder()  # Create an instance of LabelEncoder
        binary_data[column] = binary_data[column].astype(str)  # Ensure data is in string format
        binary_data[column] = encoder.fit_transform(binary_data[column])  # Transform categorical values to integers

# Fill missing values with 0
binary_data.fillna(0, inplace=True)

# Define a function to compute similarity matrices based on Jaccard Coefficient (JC) or Simple Matching Coefficient (SMC)
def compute_similarity_matrix(binary_df, method='JC'):
    num_rows = binary_df.shape[0]
    similarity_matrix = np.zeros((num_rows, num_rows))
    
    for i in range(num_rows):
        for j in range(i, num_rows):
            # Extract binary vectors for the i-th and j-th rows
            vec1 = binary_df.iloc[i].values
            vec2 = binary_df.iloc[j].values
            
            # Calculate counts for matching and non-matching binary values
            a11 = np.sum((vec1 == 1) & (vec2 == 1))
            a00 = np.sum((vec1 == 0) & (vec2 == 0))
            a01 = np.sum((vec1 == 0) & (vec2 == 1))
            a10 = np.sum((vec1 == 1) & (vec2 == 0))
            
            # Compute similarity based on the chosen method
            if method == 'JC':
                denom = a01 + a10 + a11
                similarity = a11 / denom if denom != 0 else np.nan
            elif method == 'SMC':
                denom = a00 + a01 + a10 + a11
                similarity = (a11 + a00) / denom if denom != 0 else np.nan
            else:
                raise ValueError("Unsupported similarity method")
            
            # Populate the similarity matrix
            similarity_matrix[i, j] = similarity
            similarity_matrix[j, i] = similarity
    
    return similarity_matrix

# Compute similarity matrices for both Jaccard Coefficient and Simple Matching Coefficient
jaccard_matrix = compute_similarity_matrix(binary_data, method='JC')
matching_matrix = compute_similarity_matrix(binary_data, method='SMC')

# Define a function to compute cosine similarity matrix
def compute_cosine_similarity_matrix(df):
    num_rows = df.shape[0]
    cosine_matrix = np.zeros((num_rows, num_rows))
    
    for i in range(num_rows):
        for j in range(i, num_rows):
            vec1 = df.iloc[i].values.reshape(1, -1)
            vec2 = df.iloc[j].values.reshape(1, -1)
            cosine_sim = cosine_similarity(vec1, vec2)[0][0]
            cosine_matrix[i, j] = cosine_sim
            cosine_matrix[j, i] = cosine_sim
    
    return cosine_matrix

# Calculate cosine similarity matrix
cosine_matrix = compute_cosine_similarity_matrix(binary_data)

# Plot heatmaps for Jaccard Coefficient, Simple Matching Coefficient, and Cosine Similarity
plt.figure(figsize=(12, 8))
sns.heatmap(jaccard_matrix, annot=True, cmap='viridis', cbar=True, square=True, fmt='.2f')
plt.title('Heatmap of Jaccard Coefficient')
plt.show()

plt.figure(figsize=(12, 8))
sns.heatmap(matching_matrix, annot=True, cmap='viridis', cbar=True, square=True, fmt='.2f')
plt.title('Heatmap of Simple Matching Coefficient')
plt.show()

plt.figure(figsize=(12, 8))
sns.heatmap(cosine_matrix, annot=True, cmap='viridis', cbar=True, square=True, fmt='.2f')
plt.title('Heatmap of Cosine Similarity')
plt.show()
```
