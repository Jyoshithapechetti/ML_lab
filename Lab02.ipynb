```python
#A1
import pandas as pd
import numpy as np
df = pd.read_csv('Lab Session Data.csv')
df.tail()
```

```python
# Drop columns with 'Unnamed' in their header (often created when saving data with extra index columns)
df = df.drop(df.filter(like='Unnamed').columns, axis=1)

# Drop specific columns that are not needed for the analysis
df = df.drop(['Candy', 'Mango', 'Milk'], axis=1)

# Display the first few rows of the cleaned DataFrame to verify changes
df.head()

# Extract feature columns for the matrix A and target column C
# Features used for the model
A = df[['Candies (#)', 'Mangoes (Kg)', 'Milk Packets (#)']]

# Target variable (Payment amount) to be predicted
C = df['Payment (Rs)']

# Print Matrix A to verify the extracted features
print("Matrix A:")
print(A)

# Print Matrix C to verify the target values
print("\nMatrix C:")
print(C)
```

```python
# Dimensionality of the vector space
dim = A.shape[1]
print(f"Dimensionality of the vector space is: {dim}")

# Number of vectors
num = A.shape[0]
print(f"Number of vectors is: {num}")

# Rank of Matrix A
rank_A = np.linalg.matrix_rank(A)
print(f"Rank of Matrix A: {rank_A}")

# Computing the pseudo-inverse of A
A_pseudo_inv = np.linalg.pinv(A)


```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score

def load_data():
    # Extract features (A) and target (C) from the DataFrame
    A = df[['Candies (#)', 'Mangoes (Kg)', 'Milk Packets (#)']]
    C = df['Payment (Rs)']
    return A, C

def compute_labels(C):
    # Generate labels based on the condition
    return np.where(C > 200, 'RICH', 'POOR').flatten()

#Main function to execute the classification process.
def main():
    # Load the data
    A, C = load_data()
    
    # Compute the target labels
    y = compute_labels(C)
    
    # Features for model training
    X = A
    
    # Split the data into training and testing sets
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.3, random_state=42, stratify=y
    )
    # Initialize and train the logistic regression model
    model = LogisticRegression(max_iter=1000)
    model.fit(X_train, y_train)
    
    # Make predictions on the test set
    y_pred = model.predict(X_test)
    
    # Print classification report and accuracy score
    print("Classification Report:")
    print(classification_report(y_test, y_pred, zero_division=0))
    print("Accuracy Score:")
    print(accuracy_score(y_test, y_pred))

if __name__ == "__main__":
    # Execute the main function
    main()
```
